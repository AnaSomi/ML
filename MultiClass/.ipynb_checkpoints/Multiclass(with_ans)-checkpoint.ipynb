Переведите типы монстров в 0, 1, 2

```python
def f(ans):
    if ans == 'Ghoul':
        return 0
    elif ans == 'Goblin':
        return 1
    else:
        return 2
```

```python
df.type = df.type.apply(f)
```

Равномерны ли классы?

_Ответ_ : Практически да

```python
df.groupby(by=["type"]).id.count()
```

```python
from sklearn import preprocessing
temp = np.unique(df.color).tolist()
le = preprocessing.LabelEncoder()
le.fit(temp)
df.color = le.fit_transform(df.color)
df.head()
```

Постройте распределение объектов по всем парам числовых признаков

Какие два признака лучше всего отделяют классы друг от друга?

_Ответ_ : has_soul и rotting_flesh

```python
fig, axs = plt.subplots(4, 3, figsize=(18,18))
temp = df.columns.tolist()
k = 0
l = 0
for i in temp:
    if i == 'id' or i == 'color' or i == 'type':
        continue
    for j in temp:
        if j == 'id' or j == 'color' or j == 'type' or i == j:
            continue
        sns.scatterplot(x=i, y=j, hue="type", data=df, palette=['green','blue','red'], ax=axs[k][l])
        l += 1
    k += 1
    l = 0
```

(Задание за доп балл) Придумайте новые признаки

```python
temp = df.columns.tolist()
k = 0
l = 0
for i in temp:
    if i == 'id' or i == 'color' or i == 'type':
        continue
    for j in temp:
        if l <= k or j == 'id' or j == 'color' or j == 'type' or i == j:
            if l <= k:
                l += 1
            continue
        s1 = i[:i.find('_')]
        s2 = i[i.find('_') + 1 : ]
        s = s2
        if s1 != 'length' and s1 != 'has' and s1 != 'rotting':
            s = s1
        
        s1 = j[:j.find('_')]
        s2 = j[j.find('_') + 1 : ]
        t = s2
        if s1 != 'length' and s1 != 'has' and s1 != 'rotting':
            t = s1
        df[s + '_' + t] = df[i] * df[j]
        l += 1
    k += 1
    l = 0
```
